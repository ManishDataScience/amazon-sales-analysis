```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Amazon Sale Report (1).csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#drop unrelated/blank columns
df.drop(['New','PendingS'], axis=1, inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Amount'].fillna(df['Amount'].mean(), inplace=True)
```

```python
# # Filling missing values in categorical column using mode (most frequent value)

df['currency'].fillna(df['currency'].mode()[0], inplace=True)
```

```python
df['ship-city'].fillna('Unknown', inplace=True)
df['ship-state'].fillna('Unknown', inplace=True)
df['ship-postal-code'].fillna('Unknown', inplace=True)
df['ship-country'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['fulfilled-by'].fillna(df['fulfilled-by'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.columns
```

```python
# change data type
df['ship-postal-code'] = pd.to_numeric(df['ship-postal-code'], errors='coerce').fillna(0).astype(int)


```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Month_Name'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day_name()
```

```python
# 👉 Ye sabhi orders ka total paisa add karta hai

total_sales = df["Amount"].sum()
print(total_sales)
```

```python
print("Total Sales (in Millions):", round(total_sales/1000000, 2), "M")
```

```python
ax = sns.countplot(x='Size', data=df)

for i in ax.containers:
    ax.bar_label(i)

plt.title("Order Count by Size")
plt.show()
```

```python
# 👉 Kaunsa product sabse zyada bik raha hai

cat = df["Category"].value_counts().head(5)
```

```python
cat.plot(kind="bar")
plt.title("Top Categories by Order Count")

plt.show()

```

```python
# 👉 Kaunsi state se sabse zyada revenue aa raha

state_sales = df.groupby("ship-state")["Amount"].sum().sort_values(ascending=False)
print(state_sales.head())
```

```python
state_sales.head(10).plot(kind="bar")
plt.title("Top 10 States by Sales")
plt.show()
```

```python
status = df["Status"].value_counts()
print(status)
```

```python
top_status = df['Status'].value_counts().head(2)

top_status.plot(kind='pie', autopct='%1.1f%%')
plt.title("Delivered vs Cancelled")
plt.ylabel("")
plt.show()
```

```python
# Amazon Marchant Unknown


fulfil = df["Fulfilment"].value_counts()
print(fulfil)
```

```python
fulfil.plot(kind="bar")
plt.title("Fulfilment Type")
plt.show()
```

```python
# 👉 Kaunsa size zyada demand me hai

size = df["Size"].value_counts()
print(size)
```

```python
top5_size_rev = df.groupby('Size')['Amount'].sum().sort_values(ascending=False).head(5)

top5_size_rev.plot(kind='bar')
plt.title("Top 5 Sizes by Revenue")
plt.xlabel("Size")
plt.ylabel("Revenue")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

monthly = df.groupby(df["Date"].dt.month)["Amount"].sum()
print(monthly)
```

```python
monthly_sales = df.groupby('Month_Name')['Amount'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Revenue")
plt.show()
```

```python
day_sales = df.groupby('Day')['Amount'].sum()

day_sales.plot(kind='bar')
plt.title("Sales by Day")
plt.show()
```
